#### Author and narrator

```python
df['author'] = df['author'].str.split(":").str[1]
df['narrator'] = df['narrator'].str.split(":").str[1]
```

```python
df.head()
```

```python
def add_space_before_second_upper(s):
    return re.sub(pat = r'(?<=\w)(?=[A-Z][a-z])', repl = ' ', s)
```

```python
df['author'] = df['author'].apply(add_space_before_second_upper).str.strip()
df['narrator'] = df['narrator'].apply(add_space_before_second_upper).str.strip()
```
